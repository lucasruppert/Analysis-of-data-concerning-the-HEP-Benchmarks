```python
from matplotlib.legend_handler import HandlerBase

class HandlerCenteredVerticalLines(HandlerBase):
    def __init__(self, spacing=0.1):
        self.spacing = spacing
        super().__init__()

    def create_artists(self, legend, orig_handle, x0, y0, width, height, fontsize, trans):
        lines = []
        n_lines = len(orig_handle)
        total_height = height - (n_lines - 1) * self.spacing * height
        line_spacing = total_height / n_lines
        for i, line in enumerate(orig_handle):
            y = y0 + height - (i * (line_spacing + self.spacing * height))
            x_mid = x0 + width / 2
            if isinstance(line, tuple):
                lines.append(plt.Line2D([x_mid - width / 4, x_mid + width / 4], [y, y],
                                         color=line[0].get_color(),
                                         linestyle=line[0].get_linestyle(),
                                         linewidth=line[0].get_linewidth(),
                                         transform=trans))
            else:
                lines.append(plt.Line2D([x_mid - width / 4, x_mid + width / 4], [y, y],
                                         color=line.get_color(),
                                         linestyle=line.get_linestyle(),
                                         linewidth=line.get_linewidth(),
                                         transform=trans))
        return lines
```

```python
def optimization(renew_share: float):
    times_high = df[df['Renewable share of load'] >= renew_share]['date'].to_numpy()
    times_low = df[df['Renewable share of load'] < renew_share]['date'].to_numpy()

    #plt.plot(times_low, np.zeros(len(times_low)) +  1000, '.')
    #plt.plot(times_high, np.zeros(len(times_high)) +  3000, '.')
    #plt.show()

    hours_high = len(times_high) / 4
    hours_low = len(times_low) / 4
    hours_total = len(df['date']) / 4

    days_high = hours_high / 24
    days_low = hours_low / 24
    days_total = hours_total / 24

    #print(hours_total, hours_high + hours_low)
    #print(hours_high , hours_low)
    #print(hours_total / 24)
    hs = ((hours_low * 2500 + hours_high * 3610.6 ) * 12) / hours_total

    return renew_share, hs, hs * 100/33300, days_high, days_low, (days_high / days_total) * 100, (days_low / days_total) * 100
```

```python
def optimization_data(renew_shares: np.array, hs_low, power_low, hs_high, power_high, savedata = None, freq_low = None):
    energy = np.array([])
    hs = np.array([])
    days_low = np.array([])
    days_high = np.array([])

    for renew_share in renew_shares:
        times_high = df[df['Renewable share of load'] >= renew_share]['date'].to_numpy()
        times_low = df[df['Renewable share of load'] < renew_share]['date'].to_numpy()

        sec_high = (len(times_high) / 4) * 3600
        sec_low = (len(times_low) / 4) * 3600
        sec_total = (len(df['date']) / 4) * 3600

        days_high = np.append(days_high, (len(times_high) / 4) / 24)
        days_low = np.append(days_low, (len(times_low) / 4) / 24)
        
        hs = np.append(hs, ((sec_low * hs_low + sec_high * hs_high ) * 12) / sec_total)
        energy = np.append(energy, ((sec_low * power_low + sec_high * power_high ) * 12))

    
    if savedata != None:
        pd.DataFrame({1 : renew_shares, 2 : hs, 3 : (energy / 3.6e9), 4 : days_high, 5 : days_low}).to_latex(
            savedata,
            header=['renew. energy share to switch clock-frequencies', 'median HS23', 'energy [MWh]', 'days with 2700 MHz', f'days with {freq_low} MHz'], 
            float_format="{:.1f}".format,
            index=False,
            column_format='|p{4cm}|p{2cm}|p{2cm}|p{2cm}|p{2cm}|',
        )
        

    return renew_shares, hs, energy


def interpolate_x(x_values, y_values, y_target):
    """
    Berechnet den x-Wert fÃ¼r einen gegebenen y-Wert durch lineare Interpolation.

    :param x_values: Liste oder numpy-Array mit x-Werten (muss aufsteigend sein).
    :param y_values: Liste oder numpy-Array mit y-Werten.
    :param y_target: Der y-Wert, fÃ¼r den der x-Wert berechnet werden soll.
    :return: Interpolierter x-Wert oder None, falls y_target auÃŸerhalb der Daten liegt.
    """
    for i in range(len(y_values) - 1):
        # ÃœberprÃ¼fen, ob y_target im Intervall liegt
        if y_values[i] <= y_target <= y_values[i + 1] or y_values[i] >= y_target >= y_values[i + 1]:
            # Lineare Interpolation
            x1, y1 = x_values[i], y_values[i]
            x2, y2 = x_values[i + 1], y_values[i + 1]
            x_target = x1 + (y_target - y1) / (y2 - y1) * (x2 - x1)
            return x_target
    return None  # Falls y_target auÃŸerhalb der y-Werte liegt
```

```python
def def_calc_total_co2(percent, PS_low, PS_high):
    def calc_total_co2(row):
        if row['Renewable share of generation'] > percent:
            return (0.25 * PS_high) * row['gco2eq_perkWh'] * 12
        else:
            return (0.25 * PS_low) * row['gco2eq_perkWh'] * 12
    
    return calc_total_co2
```

```python
def calc_totalCO2(percents: list[float], PS_low, PS_high):
    def co2(percent):
        return DF.apply(def_calc_total_co2(percent, PS_low, PS_high), axis=1).sum()#np.sum(DF.apply(def_calc_total_co2(percent, PS_low, PS_high), axis=1).to_numpy())

    totalco2 = np.array([])
    for i in percents:
        totalco2 = np.append(totalco2, co2(i))
    
    return totalco2
```
